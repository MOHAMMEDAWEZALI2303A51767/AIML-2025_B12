```python
people=['Ann','suraj','maaz','siduu']
age=[12,33,44,56]
weight=[22,55,85,94]
height=[160,156,123,146]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()
```

```python
plt.figure(figsize=(15,9))
plt.title("People's weight in kgs",fontsize=19,
          fontstyle='italic')
plt.bar(x=people,height=weight, width=0.55,
        color='blue',edgecolor='k',alpha=0.55)

plt.xlabel("People",fontsize=13)
plt.ylabel("Weight (in kgs)",fontsize=13)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(8,7))
plt.hist(weight,color='black',edgecolor='k', alpha=0.75,bins=5)

plt.title("Histogram of patient weight",fontsize=14)
plt.xlabel("Weight in kgs",fontsize=13)
plt.show()
```
